# Cap set problem

The cap set problem consists in finding a collection of vectors in $F_2^n$ such that no three of them add up to the all-zero vector (considering arithmetic modulo 3). We aim at finding a large collection of vectors satisfying such property.

We define the *score* as the size of the collection. For dimension $n=9$, AlphaEvolve improved the score from $1082$ to $1084$. The cap set found by AlphaEvolve is given below.

For more context on the problem, see:

[1] J. Grochow. "New applications of the polynomial method: the cap set conjecture and beyond." American Mathematical Society, 56, 29–64 (2019). 

[2] T. Tao and V. H. Vu. "Additive Combinatorics." Vol. 105 (Cambridge Univ. Press, 2006).

```python
#@title Verification

def check_cap_set_validity(
    cap_set: set[tuple[int, ...]],
    dimension: int,
) -> None:
  """Raises an error if the cap set is not valid.

  Checking the cap set property naively takes O(num_elements^3 dimension) time,
  where num_elements is the size of the given cap set. This function implements
  a faster check that runs in O(num_elements^2 dimension).

  Args:
    cap_set: A set of vectors of length `dimension` with elements in {0, 1, 2}.
    dimension: The dimension of the vectors in the cap set.
  """
  for element in cap_set:
    if (
        not isinstance(element, tuple)  # Wrong type.
        or len(element) != dimension  # Wrong length.
        or not all(isinstance(x, int) for x in element)  # Wrong type.
        or any(x < 0 or x >= 3 for x in element)  # Wrong values.
    ):
      raise ValueError(
          f"Invalid element {element} found for the {dimension}-dimensional cap "
          f"set. Each element must be a tuple of {dimension} integers in the "
          f"range [0, 2]."
      )

  num_elements = len(cap_set)

  # Convert the cap set to an array of integers.
  cap_set_np = np.array(
      list(cap_set), dtype=np.int32
  )  # Shape (num_elements, dimension).
  powers = 3 ** np.arange(dimension - 1, -1, -1)  # Shape (power,).
  cap_set_as_int = np.einsum(
      'np,p->n', cap_set_np, powers
  )  # Shape (num_elements,).

  # Starting from the empty set, we iterate through the vectors in `cap_set` one
  # by one and check that the vector can be inserted into the set without
  # violating the defining property of cap sets. To make this check fast we
  # maintain an array `invalid` indicating, for each element of Z_3^n, whether
  # that element can be inserted into the growing set without violating the cap
  # set property.
  invalid = np.zeros(shape=(3 ** dimension,), dtype=np.bool_)
  for idx_i, (new_vector, new_vector_as_int) in enumerate(
      zip(cap_set_np, cap_set_as_int, strict=True)
  ):
    if invalid[new_vector_as_int]:
      raise ValueError(
          f"Invalid element {new_vector} found for the {dimension}-dimensional "
          "cap set."
      )
    if idx_i >= 1:
      invalid_idx = np.einsum(
          'np,p->n',
          (
              - np.array(list(cap_set_np[:idx_i]), dtype=np.int32)
              - new_vector[None]
          ) % 3,
          powers,
      )
      invalid[invalid_idx] = True
    invalid[new_vector_as_int] = True

  print(
      f"{dimension}-dimensional cap set of size {num_elements} is valid."
  )


check_cap_set_validity(
    set(tuple(x) for x in discovered_cap_set), dimension=9
)
```
