# Extremal girth 5 problem
The goal of this problem is to find, for any given $n$, a simple $n$-node graph that has as many edges as possible but doesn't contain 3-cycles or 4-cycles. The *score*, for any size $n$, is the maximum number of edges subject to the no-short-cycles constraint. The best lower bounds were found in [Goedgebeur, Jooken, Joret, Van den Eede, 2025](https://arxiv.org/abs/2508.05562).

```python
#@title Verification

def check_graph_validity(n: int, edges: list[tuple[int, int]]) -> bool:
  """Checks that the graph with n nodes and given edges has no 3-cycles or 4-cycles."""
  edge_set = set()
  neighbors = [set() for _ in range(n)]

  for edge in edges:
    if len(edge) != 2:
      print(f'Graph check failed: invalid edge {edge}!')
      return False
    u, v = tuple(edge)
    if not (u in range(n) and v in range(n) and u != v):
      print(f'Graph check failed: invalid edge {edge}!')
      return False
    sorted_edge = tuple(sorted((u, v)))
    if sorted_edge in edge_set:
      print(f'Graph check failed: Edge {sorted_edge} was repeated!')
      return False

    edge_set.add(sorted_edge)
    neighbors[u].add(v)
    neighbors[v].add(u)

  # Check for 3-cycles: For every edge (u, v), check they have no common neighbor w.
  for u, v in edge_set:
    common_neighbors = neighbors[u].intersection(neighbors[v])
    if common_neighbors:
      w = next(iter(common_neighbors))
      print(f'Graph check failed: 3-cycle {(u, v, w)} detected!')
      return False

  # Check for 4-cycles: For every node u, and for all neighbors v of u, store the neighbors of v (except u) in a dictionary, called `distance_2_neighbors`.
  # If we reach the same node twice at distance 2, using distinct middle nodes v and v', we have found a 4-cycle!
  for u in range(n):
    distance_2_neighbors = dict()
    for v in neighbors[u]:
      for w in neighbors[v]:
        if w == u:
          continue
        if w in distance_2_neighbors:
          v_prime = distance_2_neighbors[w]
          print(f'Graph check failed: 4-cycle {(u, v_prime, w, v)} detected!')
          return False
        distance_2_neighbors[w] = v

  print(
      f'Verified a graph with {n} nodes and {len(edge_set)} edges without 3- or'
      ' 4-cycles.'
  )
  return True


for n, edges in ae_graphs.items():
  check_graph_validity(n, edges)
```
