# Third autocorrelation inequality

For any function $f:\mathbb{R} \rightarrow \mathbb{R}$, define the *autoconvolution* of $f$, written $f*f$, as
$$f*f (t) := \int_\mathbb{R} f(t-x) f(x)\ dx.$$

Let $C_3$ be the largest constant satisfying
$$ \max_{-1/2 \leq t \leq 1/2} \left|f*f(t)\right| \geq C_3 \left(\int_{-1/4}^{1/4} f(x)\ dx\right)^2$$
for any function $f: \mathbb{R} \rightarrow \mathbb{R}$. Then there is a step function that gives the upper bound $C_3 \leq 1.4993$ ([Matolcsi and Vinuesa (2010)](https://doi.org/10.1016/j.jmaa.2010.07.030)). AlphaEvolve discovered a step function that gives a slightly better upper bound $C_3 \leq 1.47$.
